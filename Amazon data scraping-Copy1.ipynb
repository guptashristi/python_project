```python
#importing autoscrapper 
from autoscraper import AutoScraper 
```

```python
#the URL that we want to scrap
amazon_url="https://www.amazon.in/s?k=iPhones"

#mentioning the list of items that we need
wanted_list=["₹60,999","Apple iPhone 13 (128GB) - Pink","12,851"] 
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)  
```

```python
#Price and the Title
scraper.set_rule_aliases({'rule_ecjl':'Title','rule_ikup':'Price'})
scraper.keep_rules(['rule_ecjl','rule_ikup'])
scraper.save('amazom-search')
```

```python
results= scraper.get_result_similar('https://www.amazon.in/s?k=mi+phone+under+10000+to+15000',group_by_alias= True)
```
